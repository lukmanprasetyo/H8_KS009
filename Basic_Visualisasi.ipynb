## **Pie Charts**

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))

df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                        	figsize=(5, 6),
                          autopct='%1.1%%',
                          startangle=90,
                          shadow=True,
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
df_continents['Total'].plot(kind='pie',
                        	figsize=(5, 6),
                        	autopct='%1.4f%%' # add in percentages   
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
# 2. basic pie chart + startangle

df_continents['Total'].plot(kind='pie',
                        	figsize=(5, 6),
                        	startangle=-113  # start angle 90° (Africa) # default starts from x-axis, counter clockwise
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
# 3. basic pie chart + shadow

df_continents['Total'].plot(kind='pie',
                        	figsize=(5, 6),
                        	shadow=True   	# add shadow
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
# 4. pie chart (3) + labels + legend

df_continents['Total'].plot(kind='pie',
                        	figsize=(15, 6),
                        	autopct='%1.1f%%',
                        	startangle=90,	 
                        	shadow=True,  	 
                        	labels=None   	# turn off labels on pie chart
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]')

plt.axis('equal') # Sets the pie chart to look like a circle.

# add legend
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
# 5. pie chart (4) + pctdistance + y coordinate on title

df_continents['Total'].plot(kind='pie',
                        	figsize=(15, 6),
                        	autopct='%1.1f%%',
                        	startangle=90,	 
                        	shadow=True,  	 
                        	labels=None,  	 
                        	pctdistance=1.12   # the ratio between the center of each pie slice and the start of the text generated by autopct
                        	)

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)
# plt.title('Immigration to Canada by Continent [1980 - 2013]')

plt.axis('equal') # Sets the pie chart to look like a circle.

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
# 6. pie chart (5) + colors
# colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
colors_list = ['pink', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'gold']

df_continents['Total'].plot(kind='pie',
                        	figsize=(15, 6),
                        	autopct='%1.1f%%',
                        	startangle=90,	 
                        	shadow=True,  	 
                        	labels=None,  	 
                        	pctdistance=1.12,
                        	colors=colors_list  # add custom colors
                        	)

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)

plt.axis('equal') # Sets the pie chart to look like a circle.

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
# 7. pie chart (6) + explode , like the one on the module
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0.2, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

df_continents['Total'].plot(kind='pie',
                        	figsize=(15, 6),
                        	autopct='%1.1f%%',
                        	startangle=90,	 
                        	shadow=True,  	 
                        	labels=None,  	 
                        	pctdistance=1.12,
                        	colors=colors_list,
                        	explode=explode_list # 'explode' lowest 3 continents
                        	)


plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)

plt.axis('equal') # Sets the pie chart to look like a circle.

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
# 8. pie chart (7) + modify explode argument
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
# explode_list = [0.1, 0, 0, 0, 0.05, 0.15] # ratio for each continent with which to offset each wedge.
# explode_list = [0.68, 0, 0, 0, 0.55, 0.25] # how far from the center each wedge
explode_list = [0, 0.4, 0.3, 1.2, 0, 0]

df_continents['Total'].plot(kind='pie',
                        	figsize=(15, 6),
                        	autopct='%1.1f%%',
                        	startangle=90,	 
                        	shadow=True,  	 
                        	labels=None,  	 
                        	pctdistance=1.12,
                        	colors=colors_list,
                        	explode=explode_list # 'explode' lowest 3 continents
                        	)


plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)

plt.axis('equal') # Sets the pie chart to look like a circle.

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Bubble Plots

```python
df_can_t = df_can[years].transpose()

df_can_t.index = map(int, df_can_t.index)

df_can_t.index.name = 'Year'

df_can_t.reset_index(inplace=True)

df_can_t.head()
```

```python
norm_china = (df_can_t['China'] - df_can_t['China'].min()) / (df_can_t['China'].max() - df_can_t['China'].min())
norm_india = (df_can_t['India'] - df_can_t['India'].min()) / (df_can_t['India'].max() - df_can_t['India'].min())
```

```python
ax0 = df_can_t.plot(kind='scatter',
                	x='Year',
                	y='China',
                	figsize=(14, 8),
                	alpha=0.5,              	# transparency
                	color='green',
                	s=norm_china * 1300 + 10,  # pass in weights
                	xlim=(1975, 2015) # label untuk x-axis : minimum dan maxiimu
               	)

ax1 = df_can_t.plot(kind='scatter',
                	x='Year',
                	y='India',
                	alpha=0.5,
                	color="blue",
                	s=norm_india * 1300 + 10,
                	ax = ax0
               	)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from China and India from 1980 - 2013')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
