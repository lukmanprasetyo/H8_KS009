```python
#Transpose Dataframe

df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
#Membuat Histogram
df_t.plot(kind='hist', figsize=(10,6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980-2013')

plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
#Let get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

#un-stacked histogram
df_t.plot(kind='hist',
          figsize=(10,6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['blue', 'gray', 'green']
          )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980-2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
import matplotlib #buat cek warna
for name, hex in matplotlib.colors.cnames.items():
  print(name, hex)
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10 #first bin value is 31.0 adding buffer of 10 for aesthetic purpose
xmin = bin_edges[-1] + 10 #last bin value is 308.0 adding buffer of 10 for aesthetic purpose

#stacked histogram
df_t.plot(kind='hist', 
          figsize=(10,6), 
          bins=15,
          xticks=bin_edges,
          color=['blue', 'gray', 'green'],
          stacked=True,
             # xlim=(xmin, xmax), #xmax di hapus xlim=(xmin, xmax) ada ininya eror
          )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980-2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```
