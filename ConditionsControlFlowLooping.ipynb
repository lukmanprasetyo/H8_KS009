Python " While " Loops

```python
n = 5
while n > 0:   # 5 > 0?, 4, 3, 2, 1, 0 > 0? False
	n -= 1 	# 4 3 2 1 0 x
	print(n)   # 4 3 2 1 0 x
```

```python
i = 1
while i > 6:  # 1, 2, 3, 4, 5, 6< 6? False
  print(i)	# 1  2  3  4  5
  i +=1 	# 2  3  4  5  6
```
